2. Write a function that will plot the ranking of a given roller coaster over time as a line. Your function should take a roller coaster's name and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as the roller coaster name and the wood ranking DataFrame. What issue do you notice? Update your function with an additional argument to alleviate the problem, and retest your function.

```python
# 2
# Create a function to plot rankings over time for 1 roller coaster

def plot_ranking_over_time(name, df, park):
    x = df['Year of Rank'][(df.Name == name) & (df.Park == park)].values
    y = df.Rank[(df.Name == name) & (df.Park == park)].values
    plt.plot(x, y, color='purple')
    ax = plt.subplot()
    ax.set_yticks(y)
    ax.invert_yaxis()
    plt.title(f'Ranking for {name} ({park}) over the years')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.show()
    
# Create a plot of El Toro ranking over time
plot_ranking_over_time('El Toro', ranking_wood, 'Six Flags Great Adventure')
```

3. Write a function that will plot the ranking of two given roller coasters over time as lines. Your function should take both roller coasters' names and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as one roller coaster name, `"Boulder Dash"` as the other roller coaster name, and the wood ranking DataFrame. What issue do you notice? Update your function with two additional arguments to alleviate the problem, and retest your function.

```python
# 3
# Create a function to plot rankings over time for 2 roller coasters
def compare_ranking_over_time(name1, park1, name2, park2, df):
    x1 = df['Year of Rank'][(df.Name == name1) & (df.Park == park1)]
    y1 = df.Rank[(df.Name == name1) & (df.Park == park1)]
    
    x2 = df['Year of Rank'][(df.Name == name2) & (df.Park == park2)]
    y2 = df.Rank[(df.Name == name2) & (df.Park == park2)]
    
    plt.plot(x1, y1, color='blue')
    plt.plot(x2, y2, color='grey')
    ax = plt.subplot()
    ax.set_yticks(y1)
    ax.invert_yaxis()
    plt.title(f'Ranking for {name1} ({park1}) x {name2} ({park2}) over the years')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend([name1, name2])
    plt.show()
    
# Create a plot of El Toro and Boulder Dash roller coasters
compare_ranking_over_time('El Toro', 'Six Flags Great Adventure', 'Boulder Dash', 'Lake Compounce', ranking_wood)
```

4. Write a function that will plot the ranking of the top `n` ranked roller coasters over time as lines. Your function should take a number `n` and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   For example, if `n == 5`, your function should plot a line for each roller coaster that has a rank of `5` or lower.
   
   Call your function with a value of `n` and either the wood ranking or steel ranking DataFrame.

```python
# 4
# Create a function to plot top n rankings over time
def top_n_ranking(n, df):
    top = df[df.Rank <= n]
    plt.figure(figsize=(10,10))
    for rc in set(top.Name):
        x = top['Year of Rank'][top.Name == rc]
        y = top.Rank[top.Name == rc]
        plt.plot(x, y)
    ax = plt.subplot()
    ax.set_yticks([i for i in range(1,n+1)])
    ax.invert_yaxis()
    plt.title(f'Top {n} roller coasters over the years')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend(top.Name, loc=2)
    plt.show()
    
```

```python
# Create a plot of top n rankings over time
top_n_ranking(5, ranking_wood)
img = mpimg.imread('thunderhead.jpg')
print('Thunderhead:')
plt.figure(figsize=(15,15))
imgplot = plt.imshow(img)
```

6. Write a function that plots a histogram of any numeric column of the roller coaster DataFrame. Your function should take a DataFrame and a column name for which a histogram should be constructed as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame and one of the column names.

```python
# 6
# Create a function to plot histogram of column values
def plot_histogram(df, column):
    if type(df[column][0]) != str:
        data = df[column].values
        plt.hist(data)
        plt.xlabel(column)
        plt.show()
    else:
        print('Invalid column.')
    
# Create histogram of roller coaster speed
plot_histogram(rc, 'speed')
# Create histogram of roller coaster length
plot_histogram(rc, 'length')
# Create histogram of roller coaster number of inversions
plot_histogram(rc, 'num_inversions')
# Create a function to plot histogram of height values
def height_hist(df, max_range):
    data = df.height.values
    plt.hist(data, range=(0,max_range))
    plt.xlabel('height')
    plt.show()
# Create a histogram of roller coaster height
height_hist(rc, 140)
```

7. Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park. Your function should take the roller coaster DataFrame and an amusement park name as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame and amusement park name.

```python
# 7
# Create a function to plot inversions by coaster at park
def inversions_per_park(df, park):
    filtered = df[df.park == park]
    filtered = filtered.sort_values('num_inversions', ascending=False)
    x = filtered.name
    y = filtered.num_inversions
    plt.figure(figsize=(10,8))
    plt.bar(x, y)
    plt.title(f'Number of inversions per Roller Coaster at {park}')
    plt.xlabel('Roller Coaster')
    plt.ylabel('Number of Inversions')
    plt.legend([park])
    ax = plt.subplot()
    ax.set_xticks(range(len(x)))
    ax.set_xticklabels(x, rotation=90)
    plt.show()
# Create barplot of inversions by roller coasters
inversions_per_park(rc, 'Six Flags Great Adventure')
```

8. Write a function that creates a pie chart that compares the number of operating roller coasters (`'status.operating'`) to the number of closed roller coasters (`'status.closed.definitely'`). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame.

```python
# 8
# Create a function to plot a pie chart of status.operating
def pie_plot_operating(df):
    operating = len(df.status[df.status == 'status.operating'])
    closed = len(df.status[df.status == 'status.closed.definitely'])
    others = len(df) - operating - closed
    statuses = [operating, closed, others]
    labels = ['Operating', 'Closed definitely', 'Others']
    plt.title('Roller Coaster status')
    plt.pie(statuses, autopct='%d%%', labels=labels)
    plt.axis('equal')
    plt.show()
    
# Create pie chart of roller coasters
pie_plot_operating(rc)
```

### What roller coaster seating type is most popular?

```python
rc_steel = rc[rc.material_type == 'Steel']
rc_wood = rc[rc.material_type == 'Wooden']
steel_joined = rc_steel.join(ranking_steel, how='inner')
wood_joined = rc_wood.join(ranking_wood, how='inner')
```

```python
seats = steel_joined.seating_type[steel_joined.Rank <= 10].value_counts()
labels = seats.index
data = seats.values
plt.figure(figsize=(10,8))
plt.pie(data, labels=labels, autopct='%d%%')
plt.show()
print('Among the top 10 steel roller coasters 80% have the "sit down" type of seat.')
```

### Do different seating types result in higher/faster/longer roller coasters?

```python
def seat_vs_par(df, column):
    seats = df.seating_type.unique()
    values = []
    for seat in seats:
        avg = df[column][df.seating_type == seat].mean()
        values.append(avg)
    plt.bar(seats, values, color='grey')
    plt.xlabel('Seating type')
    plt.ylabel(f' Average {column}')
    plt.title(f'Seating type x average {column}')
    ax = plt.subplot()
    ax.set_xticks(range(len(seats)))
    ax.set_xticklabels(seats, rotation=90)
    plt.show()
    
seat_vs_par(rc, 'height')
seat_vs_par(rc, 'speed')
seat_vs_par(rc, 'length')
print('It seems like floorless seats roller coasters are easier to implement greater height, length and speed.')
```

### Do roller coaster manufactures have any specialties (do they focus on speed, height, seating type, or inversions)?

```python
def manufacturer_focus(df, manufacturer):
    heights = df.height[df.manufacturer == manufacturer]
    seats = df.seating_type[df.manufacturer == manufacturer].value_counts()
    seat_count, seat_type = seats.values, seats.index
    speeds = df.speed[df.manufacturer == manufacturer]
    inversions = df.num_inversions[df.manufacturer == manufacturer].value_counts()
    inv_count, inv_type = inversions.values, inversions.index
    
    plt.figure(figsize=(15,10))
    plt.subplot(4,1,1)
    plt.hist(heights)
    plt.xlabel('Height')
    plt.ylabel('Count')
    plt.subplot(4,1,2)
    plt.bar(seat_type, seat_count)
    plt.xlabel('Seating Type')
    plt.ylabel('Count')
    plt.subplot(4,1,3)
    plt.hist(speeds)
    plt.xlabel('Speed')
    plt.ylabel('Count')
    plt.subplot(4,1,4)
    plt.bar(inv_type, inv_count)
    plt.xlabel('Number of Inversions')
    plt.ylabel('Count')
    plt.subplots_adjust(hspace=0.5)
    plt.show()
```

```python
manufacturer_focus(rc, 'CCI')
print('CCI seems to focus on fast roller coasters, having only one type of seat and no inversions in all its roller coasters.')
```

### Do amusement parks have any specialties?

```python
def park_focus(df, park):
    heights = df.height[df.park == park]
    seats = df.seating_type[df.park == park].value_counts()
    seat_count, seat_type = seats.values, seats.index
    speeds = df.speed[df.park == park]
    inversions = df.num_inversions[df.park == park].value_counts()
    inv_count, inv_type = inversions.values, inversions.index
    
    plt.figure(figsize=(15,10))
    plt.subplot(4,1,1)
    plt.hist(heights)
    plt.xlabel('Height')
    plt.ylabel('Count')
    plt.subplot(4,1,2)
    plt.bar(seat_type, seat_count)
    plt.xlabel('Seating Type')
    plt.ylabel('Count')
    plt.subplot(4,1,3)
    plt.hist(speeds)
    plt.xlabel('Speed')
    plt.ylabel('Count')
    plt.subplot(4,1,4)
    plt.bar(inv_type, inv_count)
    plt.xlabel('Number of Inversions')
    plt.ylabel('Count')
    plt.subplots_adjust(hspace=0.5)
    plt.show()
```

```python
park_focus(rc, 'Disneyland Park')
print('We can see that parks for children such as Disneyland focus on shorter, slower and not too wild roller coasters.')
```
